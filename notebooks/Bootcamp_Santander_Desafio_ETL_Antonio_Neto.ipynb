```python
import copy
import users

data_users = users.users
users_data = copy.deepcopy(data_users)
```

```python
import pandas as pd
df = pd.read_csv('data/userids.csv')
users_ids = df['UserID'].to_list()

def get_info(id, users):
  for user in users:
    if user['id'] == id:
      return user
  return None

read_users = [user for id in users_ids if (user := get_info(id, users_data)) is not None]
```

TRANSFORM

```python
def gerar_mensagem_ia(user):
    nome = user["name"]
    saldo = user["account"]["balance"]
    limite = user["account"]["limit"]

    return (
        f"Olá {nome}! Analisamos sua conta e você possui saldo de "
        f"R$ {saldo:.2f} com limite disponível de R$ {limite:.2f}. "
        "Temos ofertas personalizadas esperando por você!"
    )
```

```python
usuarios_transformados = []

for user in read_users:
    user_copy = copy.deepcopy(user)

    mensagem = gerar_mensagem_ia(user_copy)

    user_copy["news"].append({
        "id": 999,
        "icon": "https://cdn-icons-png.flaticon.com/512/4712/4712109.png",
        "description": mensagem
    })

    usuarios_transformados.append(user_copy)

print(usuarios_transformados[0]['news'])
```

LOAD

```python
import json

json_path = '/content/drive/MyDrive/DIO/Desafio_ETL/usuarios_enriquecidos.json'

with open(json_path, 'w', encoding='utf-8') as f:
    json.dump(usuarios_transformados, f, ensure_ascii=False, indent=2)
```
